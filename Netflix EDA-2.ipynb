# 1.Ratings distribution per value, movie, customer and rating year

```python
# Count of each rating
rating_counts = training_df['Rating'].value_counts().sort_index()

# Plotting the distribution of ratings
plt.figure(figsize=(8, 6))
rating_counts.plot(kind='bar')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Count of ratings per movie
ratings_per_movie = training_df['MovieID'].value_counts()

# check the range
min_ratings_per_movie = ratings_per_movie.min()
max_ratings_per_movie = ratings_per_movie.max()

print(f"Minimum number of ratings per movie: {min_ratings_per_movie}")
print(f"Maximum number of ratings per movie: {max_ratings_per_movie}")

# Plotting the count of ratings per movie
plt.figure(figsize=(10, 6))
ratings_per_movie.hist(bins=50, range=(0, 200000))  # Adjust the range and bins as needed
plt.title('Count of Ratings per Movie')
plt.xlabel('Number of Ratings per Movie')
plt.ylabel('Count of Movies')
plt.grid(axis='y')
plt.show()
```

As it can be seen in the above plot rating distribution per movie is skewed to left and there are a few number of movies with a exteremly high number of ratings. This possibly can lead to popularity bias in the recommender system. But we can not exclude these movies from data set

```python
# Count of ratings per user
ratings_per_customer = training_df['CustomerID'].value_counts()

# Check the range
min_ratings_per_customer = ratings_per_customer.min()
max_ratings_per_customer = ratings_per_customer.max()

print(f"Minimum number of ratings per customer: {min_ratings_per_customer}")
print(f"Maximum number of ratings per customer: {max_ratings_per_customer}")

# Plotting the count of ratings per user
plt.figure(figsize=(10, 6))
ratings_per_customer.hist(bins=30, range=(0, 500))  
plt.title('Count of Ratings per Customer')
plt.xlabel('Number of Ratings per Customer')
plt.ylabel('Count of Customers')
plt.grid(axis='y')
plt.show()

```

Same as the previous plot, number of ratings per user is also skewed to left. Most of the users have close to zero rating in the dataset which shows the sparsity of the data. Also there are some super users with extremely high number of ratings 

```python
# Sparsity of the dataset
total_possible_ratings = training_df['CustomerID'].nunique() * training_df['MovieID'].nunique()
actual_ratings = len(training_df)
sparsity = 1 - (actual_ratings / total_possible_ratings)
print(f"Sparsity of the dataset: {sparsity:.2%}")
```

```python
# Count of ratings per year
ratings_per_year = training_df['RatingYear'].value_counts().sort_index()

# Plotting the count of ratings per year
plt.figure(figsize=(10, 6))
ratings_per_year.plot(kind='bar')
plt.title('Count of Ratings per Year')
plt.xlabel('Year')
plt.ylabel('Count of Ratings')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 2. Average of ratings per movies

```python
# Calculate the average rating for each movie
average_rating_per_movie = training_df.groupby('MovieID')['Rating'].mean()

# Plotting the distribution of average ratings per movie
plt.figure(figsize=(10, 6))
average_rating_per_movie.hist(bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Average Ratings per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Count of Movies')
plt.grid(axis='y')
plt.show()
```
